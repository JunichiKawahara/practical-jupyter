# 3-8 クロス集計

```python
# groupby メソッドで集約する
import os
import pandas as pd
import numpy as np

base_url = 'https://raw.githubusercontent.com/practical-jupyter/sample-data/master/anime/'
anime_master_csv = os.path.join(base_url, 'anime_master.csv')
df = pd.read_csv(anime_master_csv)

grouped = df.groupby('type')
type(grouped)
```

```python
grouped.mean().round(1)
```

```python
grouped.describe().round(1).head(16)
```

```python
# 集約された複数列データの平均値
# type列とepisodes列に基づいてデータを集約し、平均を求める。
df.groupby(['type', 'episodes']).mean().round(1).head(20)
```

```python
# pivot_tableメソッドで集約する
df.pivot_table(index='type', aggfunc=np.mean).round(1)
```

```python
df.pivot_table(index=['type', 'episodes'], aggfunc=np.mean).round(1).head(20)
```

## クロス集計する
これまではおもにtype列に対して集計してきた。ここからはtype列とgenre列の2項目によるクロス集計を行う。
まず、次の前処理を実施する。

* カンマ区切りで入力されていたgenre列をユニークにして取り出す。
* 元のDataFrameから↑のデータを抽出する。
* ↑を結合する。

```python
df['genre'].head(10)
```

```python
# genre列のカンマ区切りのデータを分割する
genres = df['genre'].map(lambda x: x.split(','))

# numpy.array にして2次元から1次元のデータに変換する
ser = pd.Series(np.hstack(genres.values))

# ユニークにする
unique_genres = ser.str.strip().unique()
unique_genres.sort()
unique_genres
```

```python
# ジャンルごとのDataFrameを結合する前処理
# 指定したジャンル名をDataFrameから抽出する
def filter_df_by_genre(df, genre):
    genre_df = df.loc[df['genre'].map(lambda x: genre in x)].copy()
    genre_df['genre'] = genre
    return genre_df

# 上記の関数をすべてのジャンルに対して実行する
genre_df_list = [filter_df_by_genre(df, genre) for genre in unique_genres]
# 上記dataを結合する
df2 = pd.concat(genre_df_list)
# name列でソート
df2.sort_values('name', inplace=True)

# メンバ数が多いジャンルトップ10
top10 = df2.groupby('genre')['members'].sum().sort_values(
    ascending=False).index[:10]
# top10からデータを抽出する
df2 = df2[df2['genre'].isin(top10)]

df2.head(10)
```

```python
# 「君の名は。」の抽出
df2.loc[df2['name'] == 'Kimi no Na wa.']
```

```python
# pivot_table()メソッドを使用してクロス集計を行う。
# genre列を縦軸、type列を横軸にし、members列の合計をクロス集計する。
df2.pivot_table(index='genre', columns='type', values=['members'], aggfunc=np.sum).head()
```
